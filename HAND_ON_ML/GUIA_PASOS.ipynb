### Preparar los datos. 
Notas:
* Trabaja con copias de los datos (mantén el conjunto de datos original intacto).
* Escriba funciones para todas las transformaciones de datos que aplique, por cinco razones:
    * Para que puedas preparar fácilmente los datos la próxima vez que tengas un conjunto de datos fresco.
    * Para que pueda aplicar estas transformaciones en futuros proyectos
    * Para limpiar y preparar el conjunto de pruebas
    * Para limpiar y preparar nuevas instancias de datos una vez que la solución esté en funcionamiento.
    * Para facilitar el tratamiento de sus opciones de preparación como hiperparámetros

1. Limpieza de datos:
    * Corregir o eliminar los valores atípicos (opcional).
    * Rellenar los valores que faltan (por ejemplo, con cero, media, mediana...) o eliminar sus filas (o columnas).
2. Selección de características (opcional):
    * Eliminar los atributos que no proporcionan información útil para la tarea.
3. Ingeniería de características, cuando proceda:
    * Discretizar las características continuas.
    * Descomponer las características (por ejemplo, categóricas, fecha/hora, etc.).
    * Añadir transformaciones prometedoras de las características (por ejemplo, log(x), sqrt(x), x , etc.).
    * Agregar características en nuevas características prometedoras.
4. Escalado de características:
    * Estandarizar o normalizar las características.
    
### Lista de Modelos Prometedores.

Notas:
* Si los datos son enormes, puede querer muestrear conjuntos de entrenamiento más pequeños para poder entrenar muchos modelos diferentes en un tiempo razonable (tenga en cuenta que esto penaliza a los modelos complejos como las grandes redes neuronales o los bosques aleatorios).
* Una vez más, intente automatizar estos pasos en la medida de lo posible.
1. Entrene muchos modelos rápidos y sucios de diferentes categorías (por ejemplo, lineal, naive Bayes, SVM, Random Forest, red neuronal, etc.) utilizando parámetros estándar.
2. Medir y comparar su rendimiento.
    * Para cada modelo, utilice la validación cruzada de N-fold y calcule la media y la desviación estándar de la medida de rendimiento en los N-fold.
3. Analizar las variables más significativas para cada algoritmo.
4. Analice los tipos de errores que cometen los modelos.
    * ¿Qué datos habría utilizado un humano para evitar estos errores?
5. Realice una ronda rápida de selección e ingeniería de características.
6. Realice una o dos iteraciones rápidas más de los cinco pasos anteriores.
7. Preseleccione los tres o cinco modelos más prometedores, prefiriendo los modelos que cometen diferentes tipos de errores.

###  Afinar el sistema
Notas:
* Querrá utilizar tantos datos como sea posible para este paso, especialmente a medida que se acerque al final del ajuste fino.
* Como siempre, automatice lo que pueda.
1. Afine los hiperparámetros utilizando la validación cruzada:
    * Trate sus opciones de transformación de datos como hiperparámetros, especialmente cuando no esté seguro de ellos (por ejemplo, si no está seguro de si reemplazar los valores perdidos con ceros o con el valor de la mediana, o simplemente eliminar las filas).
    * A menos que haya muy pocos valores de hiperparámetros que explorar, prefiera la búsqueda aleatoria a la búsqueda en cuadrícula. Si el entrenamiento es muy largo, puede preferir un enfoque de optimización bayesiana (por ejemplo, utilizando priores del proceso gaussiano, como se describe por Jasper Snoek et al.).
2. Pruebe los métodos de conjunto. La combinación de sus mejores modelos suele producir un mejor rendimiento que si se ejecutan individualmente.

3. Una vez que esté seguro de su modelo final, mida su rendimiento en el conjunto de pruebas para estimar el error de generalización.

**WARNING**|
:----------------------------:|
No modifique su modelo después de medir el error de generalización: sólo empezaría a sobreajustar el conjunto de pruebas.
